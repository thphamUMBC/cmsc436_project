- \* = indicates that a wage estimate is not available
- \*\* = indicates that an employment estimate is not available
- \# = indicates a wage equal to or greater than $100.00 per hour or $208,000 per year


Anual mean income by state and occupation:


```python
tmp_result = income_df[["PRIM_STATE", "OCC_TITLE", "A_MEAN", "A_MEDIAN"]]
# the excluded states are from later analysis
tmp_result = tmp_result.loc[~tmp_result["PRIM_STATE"].isin(['GU', 'PR', 'VI'])]
def A_mean_meadian_state_occupations(states=None, occupations=None):
    result = []
    if states==None and occupations==None:
        result = tmp_result
    elif states!=None and occupations!=None:
        result = tmp_result.loc[tmp_result["PRIM_STATE"].isin(states) & tmp_result["OCC_TITLE"].isin(occupations)]
    elif states==None:
        result = tmp_result.loc[tmp_result["OCC_TITLE"].isin(occupations)]
    elif occupations==None:
        result = tmp_result.loc[tmp_result["PRIM_STATE"].isin(states)]
    if result.__len__() != 0:
        return result.replace(['*', '#', '**'], [0, 208000,  np.nan])
    return None
```

```python
test = A_mean_meadian_state_occupations(states=['AL', 'AK', 'VA'], occupations=['Compensation and Benefits Managers', 'Architects, Except Landscape and Naval'])
test
```

## House price dataset


Available states:


```python
house_state = np.array(house_price_df["State"].drop_duplicates())
state_dict = {}
# missing state in house price:
missing_state = []
for st in income_state:
    state_dict[st] = 0
for st in house_state:
    state_dict[st] += 1
for key, value in state_dict.items():
    if value == 0:
        missing_state.append(key)
missing_state
```

Therefore, Guam, Puerto Rico, and U.S. Virginia Islands are not in the list of the house price


Extract data from house_price_df

```python
def median_hprice_by_state_in_YYMM(states=None, yyyymmdd=None):
    result = []
    if states==None and yyyymmdd==None:
        # return house_price_df[["State", "RegionName", "2023-01-31"]].sort_values("State")
        result = house_price_df[["State", "RegionName", "2023-01-31"]]
    elif states!=None and yyyymmdd!=None:
        tmp_result = house_price_df[np.hstack(tup=(["State", "RegionName"],yyyymmdd))]
        result = tmp_result.loc[tmp_result["State"].isin(states)]
    elif states==None:
        result = house_price_df[np.hstack(tup=(["State", "RegionName"],yyyymmdd))]
    elif yyyymmdd==None:
        result = house_price_df[np.hstack(tup=(states, ["RegionName", "2023-01-31"]))]
    result = result.round(0)
    return result
```

```python
mtest = median_hprice_by_state_in_YYMM(states=["AK", "AL"], yyyymmdd=["2023-01-31", "2022-01-31"])
mtest
```

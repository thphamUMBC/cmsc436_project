Anual mean income by state and occupation:


```python
tmp_result = income_df[["PRIM_STATE", "OCC_TITLE", "A_MEAN", "A_MEDIAN"]]
# the excluded states are from later analysis
tmp_result = tmp_result.loc[~tmp_result["PRIM_STATE"].isin(['GU', 'PR', 'VI'])]
def A_mean_meadian_state_occupations(states=None | list, occupations=None | list) -> None|pd.DataFrame:
    '''
        @states is a list of string of state abbreviation.
        @occupations is a list of string of available occupations in the income_df.
        @return dataframe that is filtered from @states and @occupations.
        @notes: replace "*" by 0, "#" by 208000 and "**" by NAN
    '''
    result = None
    if states==None and occupations==None:
        result = tmp_result
    elif states!=None and occupations!=None:
        result = tmp_result.loc[tmp_result["PRIM_STATE"].isin(states) & tmp_result["OCC_TITLE"].isin(occupations)]
    elif states==None:
        result = tmp_result.loc[tmp_result["OCC_TITLE"].isin(occupations)]
    elif occupations==None:
        result = tmp_result.loc[tmp_result["PRIM_STATE"].isin(states)]
    if result.__len__() != 0:
        return result.replace(['*', '#', '**'], [0, 208000,  np.nan])
    return None
```

Therefore, Guam, Puerto Rico, and U.S. Virgin Islands are not in the list of the house price


Extract data from house_price_df

```python
def median_hprice_by_state_in_YYMM(states=None|list, yyyymmdd=None|list) ->pd.DataFrame:
    '''
        @states is a list of string of state abbreviation.
        @yyyymmdd is a list of string of available date in house_price_df as a form of "yyyy-mm-dd". Note that dd should be the last dates in the months.
        @return dataframe of [state, regionName, yyyymmdd] that is filtered from @states and @yyyymmdd.
        @notes: the house price is round up to 0 decimal value.
    '''
    result = None
    if states==None and yyyymmdd==None:
        result = house_price_df[["State", "RegionName", "2023-01-31"]]
    elif states!=None and yyyymmdd!=None:
        tmp_result = house_price_df[np.hstack(tup=(["State", "RegionName"],yyyymmdd))]
        result = tmp_result.loc[tmp_result["State"].isin(states)]
    elif states==None:
        result = house_price_df[np.hstack(tup=(["State", "RegionName"],yyyymmdd))]
    elif yyyymmdd==None:
        result = house_price_df[np.hstack(tup=(states, ["RegionName", "2023-01-31"]))]
    result = result.round(0)
    return result
```

## Stacked bar chart

we have: salary of every occupation in each state
        house price in every county (hp)

==> stacked bar chart

so we need: to total (sum of all salary of all occupation) and every sub bar chart(salary of each occupation)

salary of each occupation (Sa)=> sa/hp ==> each sub bar chart == affordability of housing bigger it is the more affordable
==> stack bar chart, limited to n < 4 occupations
stacked bar chart = 4 occupations in selected state / hp at every county

horizontal : county

vertical: affodability ( maximum 4 occupations)

### Supportive functions

```python
# return string with whitespace replace by newline \n
def split_then_concat_with_newline(string=str):
    return string.replace(" ", "\n")
def get_raw_counties(state=str):
    return house_price_df[house_price_df['State'] == state]['RegionName']
# get county name list with whitespace replaced by \n
def get_counties(state=str):
    raw_counties = np.array(get_raw_counties(state=state))
    return np.array([split_then_concat_with_newline(x) for x in raw_counties])

# get a list of occupation
def get_raw_occupations(state=str):
    return income_df[income_df['PRIM_STATE'] == state]['OCC_TITLE']

# get a ration between median salary and house price in a state in a dictionary
# we only care about the most recent house price '2023-01-31' house price is the choice
def get_affordability_by_state_occupations(state=str, occupations=list):
    if state == None or occupations == None:
        return None
    affordability_dict = {}
    state_hp = median_hprice_by_state_in_YYMM(states=[state], yyyymmdd=['2023-01-31'])['2023-01-31']
    for occ in occupations:
        tmp = np.array(A_mean_meadian_state_occupations(states=[state], occupations=[occ])["A_MEDIAN"])[0]
        tmparr = np.array([tmp/x*100 for x in state_hp])
        affordability_dict[occ] = np.array(tmparr)
    
    return affordability_dict
```

```python
county_options = get_raw_counties('MD')
county_value = [np.array(county_options)[0]]
sel_ct = iw.SelectMultiple(
    options=np.array(county_options),
    value=county_value,
    rows=10,
    description='Counties',
    disabled=False
)


occupation_options = get_raw_occupations('MD')
occupation_value = [np.array(occupation_options)[0]]
sel_occ = iw.SelectMultiple(
    options=np.array(occupation_options),
    value=occupation_value,
    rows=10,
    description='occupations',
    disabled=False
)



def stacked_bar_by_counties_and_occupations(sel_counties=list, sel_occupations=list, state='MD'):
    '''
        counties is np.array
        affordability is dict
    '''


    all_counties = get_raw_counties(state=state)
    filter = all_counties.isin(list(sel_counties))
    affordability = get_affordability_by_state_occupations(state=state, occupations=list(sel_occupations))
    filtered_ct = np.array(all_counties[filter])
    filtered_ct = [split_then_concat_with_newline(ct) for ct in filtered_ct]
    filtered_occ = {}
    for key, value in affordability.items():
        filtered_occ[key] = value[filter]


    width = 0.4

    fig, ax = plt.subplots()
    bottom = np.zeros(filtered_ct.__len__())

    fig.set_figwidth(20)
    fig.set_figheight(10)

    for boolean, affd in filtered_occ.items():
        p = ax.bar(filtered_ct, affd, width, label=boolean, bottom=bottom)
        bottom += affd

    ax.set_title(
        'Fig1: Test'
    )
    ax.legend(loc="upper right")

    plt.show()

iw.interactive(stacked_bar_by_counties_and_occupations, sel_counties=sel_ct, sel_occupations=sel_occ, state=iw.fixed('MD'))
```

## Heatmap

salary of each occupation (Sa)=> sa/hp ==> each sub bar chart == affordability of housing bigger it is the more affordable
==> stack bar chart, limited to n < 4 occupations

color map: affordability

```python
# source: https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html

def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw=None, cbarlabel="", **kwargs):
    """
    Create a heatmap from a numpy array and two lists of labels.

    Parameters
    ----------
    data
        A 2D numpy array of shape (M, N).
    row_labels
        A list or array of length M with the labels for the rows.
    col_labels
        A list or array of length N with the labels for the columns.
    ax
        A `matplotlib.axes.Axes` instance to which the heatmap is plotted.  If
        not provided, use current axes or create a new one.  Optional.
    cbar_kw
        A dictionary with arguments to `matplotlib.Figure.colorbar`.  Optional.
    cbarlabel
        The label for the colorbar.  Optional.
    **kwargs
        All other arguments are forwarded to `imshow`.
    """

    if ax is None:
        ax = plt.gca()

    if cbar_kw is None:
        cbar_kw = {}

    # Plot the heatmap
    im = ax.imshow(data, **kwargs)

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")

    # Show all ticks and label them with the respective list entries.
    ax.set_xticks(np.arange(data.shape[1]), labels=col_labels)
    ax.set_yticks(np.arange(data.shape[0]), labels=row_labels)

    # Let the horizontal axes labeling appear on top.
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=-30, ha="right",
             rotation_mode="anchor")

    # Turn spines off and create white grid.
    ax.spines[:].set_visible(False)

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=5)
    ax.tick_params(which="minor", bottom=False, left=False)

    return im, cbar


def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=("black", "white"),
                     threshold=None, **textkw):
    """
    A function to annotate a heatmap.

    Parameters
    ----------
    im
        The AxesImage to be labeled.
    data
        Data used to annotate.  If None, the image's data is used.  Optional.
    valfmt
        The format of the annotations inside the heatmap.  This should either
        use the string format method, e.g. "$ {x:.2f}", or be a
        `matplotlib.ticker.Formatter`.  Optional.
    textcolors
        A pair of colors.  The first is used for values below a threshold,
        the second for those above.  Optional.
    threshold
        Value in data units according to which the colors from textcolors are
        applied.  If None (the default) uses the middle of the colormap as
        separation.  Optional.
    **kwargs
        All other arguments are forwarded to each call to `text` used to create
        the text labels.
    """

    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()

    # Normalize the threshold to the images color range.
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.

    # Set default alignment to center, but allow it to be
    # overwritten by textkw.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)

    # Get the formatter in case a string is supplied
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)

    # Loop over the data and create a `Text` for each "pixel".
    # Change the text's color depending on the data.
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            kw.update(color=textcolors[int(im.norm(data[i, j]) > threshold)])
            text = im.axes.text(j, i, valfmt(data[i, j], None), **kw)
            texts.append(text)

    return texts
```

```python
county_options = get_raw_counties('MD')
county_value = [np.array(county_options)[0]]
sel_ct = iw.SelectMultiple(
    options=np.array(county_options),
    value=county_value,
    rows=10,
    description='Counties',
    disabled=False
)


occupation_options = get_raw_occupations('MD')
occupation_value = [np.array(occupation_options)[0]]
sel_occ = iw.SelectMultiple(
    options=np.array(occupation_options),
    value=occupation_value,
    rows=10,
    description='occupations',
    disabled=False
)

def heatmeap_by_counties_occupations(sel_counties, sel_occupations, state):
    all_counties = get_raw_counties(state=state)
    filter = all_counties.isin(list(sel_counties))
    affordability = get_affordability_by_state_occupations(state, list(sel_occupations))
    filtered_ct = np.array(all_counties[filter])
    filtered_afd = []
    filtered_occ = []
    for key, value in affordability.items():
        filtered_afd.append(value[filter])
        filtered_occ.append(key)

    filtered_afd = np.array(filtered_afd)
    fig, ax = plt.subplots()

    im, cbar = heatmap(filtered_afd, filtered_occ, filtered_ct, ax=ax,
                   cmap='YlGnBu', cbarlabel="affordability [%]")
    texts = annotate_heatmap(im, valfmt="{x:.1f}")

    fig.tight_layout()  
    plt.show()

iw.interactive(heatmeap_by_counties_occupations, sel_counties=sel_ct, sel_occupations=sel_occ, state=iw.fixed('MD'))
```
